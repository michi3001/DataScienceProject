# Diamant-Datensatz Beschreibung

### Übersicht

Dieser Datensatz enthält Informationen über Diamanten, einschließlich ihrer physikalischen Eigenschaften und Preise. Der Datensatz besteht aus **53.940 Zeilen** und **10 Variablen** (Merkmalen). Hier ist eine detaillierte Beschreibung der einzelnen Merkmale:

### Variablen:

- **price**  
  Preis des Diamanten in US-Dollar (von \$326 bis \$18.823).

- **carat**  
  Gewicht des Diamanten (von 0,2 bis 5,01 Karat).

- **cut**  
  Qualität des Diamantschliffs. Die Kategorien umfassen:
  - Fair
  - Good
  - Very Good
  - Premium
  - Ideal

- **color**  
  Diamant-Farbgrad, der von J (schlechteste) bis D (beste) reicht.

- **clarity**  
  Ein Maß für die Reinheit des Diamanten, wobei niedrigere Werte mehr Einschlüsse und höhere Werte klarere Diamanten darstellen. Die Kategorien umfassen:
  - I1 (schlechteste)
  - SI2
  - SI1
  - VS2
  - VS1
  - VVS2
  - VVS1
  - IF (beste)

- **x**  
  Länge des Diamanten in Millimetern (von 0 bis 10,74 mm).

- **y**  
  Breite des Diamanten in Millimetern (von 0 bis 58,9 mm).

- **z**  
  Tiefe des Diamanten in Millimetern (von 0 bis 31,8 mm).

- **depth**  
  Gesamtprozent der Tiefe, berechnet als \( \text{depth} = \frac{2 \times z}{x + y} \). Werte reichen von 43% bis 79%.

- **table**  
  Die Breite der Oberseite des Diamanten relativ zu seinem breitesten Punkt, von 43% bis 95%.


```python
if not silent:
    print(f"Datensatz bevor er transformiert wurde:\n{df.dtypes.to_frame(name='Data Type')}")
    display(df.head())

# Zeige und lösche Spalten die nur NaN Werte enthalten
colsAllMissing = df.columns[df.isnull().all()]
if not silent:
    print(f"Der Datensatz enthaelt folgende Spalten die nie Werte haben: {colsAllMissing.values}")
df = df.drop(columns=colsAllMissing)

# Erste Spalte löschen, da es sich um einen ID im Dokument handelt
df = df.drop(columns=df.columns[0])

# Werte casten
df['cut'] = df['cut'].astype('string')
df['color'] = df['color'].astype('string')
df['clarity'] = df['clarity'].astype('string')


print(f"Finale Struktur des Datensatzes:\n{df.dtypes.to_frame(name='Data Type')}")


```
